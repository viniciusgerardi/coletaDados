### Carregando os dados de base World Happiness no MongoDB

Importando bibliotecas

```python
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
import json
```

Adicionando dados do servidor (rodando localmente)

```python
client = MongoClient('mongodb://localhost:27017')
db=client.admin
```

Testando a conex√£o

```python
serverStatusResult=db.command("serverStatus")
pprint(serverStatusResult)
```

Listando databases existentes

```python
print(client.list_database_names())
```

Criando database

```python
# cria database
whdb = client["worldHappiness"]

# Cria collection
whcol = whdb["happiness"]
```

Importando dados para o MongoDB

```python
# Criando um dataframe com os dados do csv
dados = pd.read_csv('world-happiness-report-2021.csv')
print(dados.head)
```

```python
# Convertendo csv para json
carga = json.loads(dados.to_json(orient='records'))
```

```python
# Carregando dados
whcol.insert(carga)
```

### Verificando se os dados foram carregados

Checando databases existentes

Fazendo uma pesquisa no banco, selecionando os pa√≠ses da Am√©rica Latina
